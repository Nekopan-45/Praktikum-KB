# Diagram Pie

```python
ds = ds.dropna(subset=['processor_brand'])
```

```python
data_pie = ds['processor_brand'].value_counts().rename_axis('processor_brand').reset_index(name='brand')
```

```python
plt.figure(figsize=(7,7))
plt.pie(data_pie.brand, labels=data_pie.processor_brand, startangle=90,autopct='%.1f%%')
plt.title('Brand Processor')
plt.show()
```

Berdasarkan diagram pie diatas kita dapat melihat intel mendominasi pasar laptop dengan persentase 73,7%, dan yang memiliki paling sedikit adalah processor dari brand Qualcomm dengan persentase 0,1%.

# Scatter Plot

```python
plt.figure(figsize=(20,6))
sns.scatterplot(data = ds, x = "brand", y = "latest_price")
plt.show()
```
