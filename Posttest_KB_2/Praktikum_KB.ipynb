```python
import numpy as np
import pandas as pd
from tabulate import tabulate
```

## Tampil Dataset

```python
dataset = pd.read_csv('steam.csv')
dataset
```

## Info Dataset

```python
dataset.info()
```

## Jumlah Record

```python
jumlah_record = len(dataset.index)
print("Jumlah Record :", jumlah_record)
```

## Nama Atribut

```python
nama_atribut = '\n'.join(list(dataset.columns))
print("Nama Atribut: ", nama_atribut)
```

## Jumlah Atribut 

```python
jumlah_atribut = dataset.shape[1]
print("Jumlah Atribut:", jumlah_atribut)
```

## Tipe Objek

```python
tipe_objek = len(dataset.dtypes[dataset.dtypes == 'object'])
print("Jumlah Atribut Tipe Objek=", tipe_objek)
```

## Tipe Angka

```python
Int = len(dataset.dtypes[dataset.dtypes == 'int64'])
Float = len(dataset.dtypes[dataset.dtypes == 'float64'])
tipe_angka = Int + Float
print("Jumlah Atribut Tipe Angka=", tipe_angka)
```

## Jumlah Nilai Null

```python
jumlah_null = dataset.isnull().any().sum()
print("Jumlah Atribut Nilai Null=", jumlah_null)
```

## Tampilan Dictionary

```python
Dict = {
    "Nama Atribut" : [nama_atribut],
    "Jumlah Record" : [jumlah_atribut],
    "Tipe Objek" : [tipe_objek],
    "Tipe Angka" : [tipe_angka],
    "Jumlah Nilai Null" : [jumlah_null],
}

tabel = pd.DataFrame(Dict)
print(tabulate(tabel, headers='keys', tablefmt='psql', showindex=False))
```

## Rata-rata, Standar deviasi, Nilai minimum, Nilai maksimum, Q1, Q2, dan Q3 dari Attribute bertipe angka.

```python
value = dataset.describe()
value.rename(index = {"count": "Jumlah Data","Mean": "Rata-rata","std": "Standar Deviasi", "min": "Nilai Minimum", 
                     "25%": "Q1", "50%": "Q2", "75%": "Q3", "max": "Nilai Maksimal"},inplace = True)
value
```
