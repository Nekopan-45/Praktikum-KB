## Data Cleaning nilai null

```python
ds.isna().sum()
```

### Mengubah nilai null bertipe data bilangan bulat

```python
imp = SimpleImputer(strategy = 'most_frequent')
ds["processor_brand"] = imp.fit_transform(ds[["processor_brand"]])
ds["ram_gb"] = imp.fit_transform(ds[["ram_gb"]])
ds["discount"] = imp.fit_transform(ds[["discount"]])
ds["Touchscreen"] = imp.fit_transform(ds[["Touchscreen"]])
ds["weight"] = imp.fit_transform(ds[["weight"]])
ds["msoffice"] = imp.fit_transform(ds[["msoffice"]])
ds["reviews"] = imp.fit_transform(ds[["reviews"]])
print("Jumlah Null: ")
ds.isna().sum()
```

### Mengubah nilai null bertipe angka bilangan desimal

```python
imp = SimpleImputer(strategy = 'mean')
ds["warranty"] = imp.fit_transform(ds[["warranty"]])
ds["star_rating"] = imp.fit_transform(ds[["star_rating"]])
print("Jumlah Null: ")
ds.isna().sum()
```

### Mengubah nilai null bertipe data kategorikal

```python
imp = SimpleImputer(strategy = 'most_frequent')
ds["processor_gnrtn"] = imp.fit_transform(ds[["processor_gnrtn"]])
ds["os"] = imp.fit_transform(ds[["os"]])
ds["brand"] = imp.fit_transform(ds[["brand"]])
ds["model"] = imp.fit_transform(ds[["model"]])
ds["processor_name"] = imp.fit_transform(ds[["processor_name"]])
ds["ssd"] = imp.fit_transform(ds[["ssd"]])
ds["hdd"] = imp.fit_transform(ds[["hdd"]])
print("Jumlah Null: ")
ds.isna().sum()
```
